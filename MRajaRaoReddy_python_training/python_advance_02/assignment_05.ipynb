```python
# Importing Matplotlib library

import matplotlib.pyplot as plt
```

```python
# Department Data

departments = ["HR", "IT", "Finance"]

employees = [5, 12, 7]
```

```python
# Bar Chart

plt.bar(
    departments,
    employees
)

plt.title("Employees by Department")

plt.show()
```

```python
# Line Chart

plt.plot(
    departments,
    employees,
    marker="o"
)

plt.title("Department Trend")

plt.show()
```

```python
# Salary Data

salaries = [
    30000,
    40000,
    50000,
    60000,
    45000
]
```

```python
# Histogram

plt.hist(
    salaries
)

plt.title("Salary Distribution")

plt.show()
```

```python
# Age and Salary Data

ages = [
    25,
    30,
    28,
    35
]

salary_values = [
    30000,
    50000,
    45000,
    60000
]
```

```python
# Scatter Plot

plt.scatter(
    ages,
    salary_values
)

plt.xlabel("Age")

plt.ylabel("Salary")

plt.show()
```
