```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create Student Dataset

student_dataframe = pd.DataFrame(
    {
        "Name": ["Rahul", "Priya", "Siri", "Anuj"],
        "Marks": [70, 80, 90, 60],
        "Hours Studied": [2, 3, 5, 1]
    }
)

student_dataframe
```

```python
# Add Performance Column

student_dataframe["Performance"] = (
    student_dataframe["Marks"].apply(
        lambda marks:
        "Pass"
        if marks > 65
        else "Fail"
    )
)

student_dataframe
```

```python
# Line Chart: Hours Studied vs Marks

plt.plot(
    student_dataframe["Hours Studied"],
    student_dataframe["Marks"],
    marker="o"
)

plt.title("Hours Studied vs Marks")

plt.show()
```

```python
# Scatter Plot: Hours Studied vs Marks

plt.scatter(
    student_dataframe["Hours Studied"],
    student_dataframe["Marks"]
)

plt.xlabel("Hours Studied")

plt.ylabel("Marks")

plt.show()
```

```python
# Bar Plot: Performance vs Marks

sns.barplot(
    data=student_dataframe,
    x="Performance",
    y="Marks"
)

plt.show()
```
