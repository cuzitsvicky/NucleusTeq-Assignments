```python
# Importing Pandas library

import pandas as pd
```

```python
# Create DataFrame with Missing Values

employee_dataframe = pd.DataFrame(
    {
        "Name": ["Rahul", "Priya", "Anuj"],
        "Age": [25, None, 29],
        "Salary": [30000, 40000, None]
    }
)

employee_dataframe
```

```python
# Detect Missing Values

employee_dataframe.isnull()
```

```python
# Calculate Mean Age

mean_age = employee_dataframe["Age"].mean()

print("Mean Age:", mean_age)
```

```python
# Replace Missing Age with Mean

employee_dataframe["Age"] = (
    employee_dataframe["Age"].fillna(mean_age)
)

employee_dataframe
```

```python
# Replace Missing Salary with 0

employee_dataframe["Salary"] = (
    employee_dataframe["Salary"].fillna(0)
)

employee_dataframe
```
