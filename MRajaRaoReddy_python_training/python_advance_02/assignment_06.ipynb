```python
# Import Libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create Employee Dataset

employee_dataframe = pd.DataFrame(
    {
        "Name": ["Rahul", "Priya", "Amit", "Anuj"],
        "Age": [25, 30, 28, 35],
        "Department": ["HR", "IT", "Finance", "IT"],
        "Salary": [30000, 50000, 45000, 60000]
    }
)

employee_dataframe
```

```python
# Bar Plot

sns.barplot(
    data=employee_dataframe,
    x="Department",
    y="Salary"
)

plt.show()
```

```python
# Box Plot

sns.boxplot(
    y=employee_dataframe["Salary"]
)

plt.show()
```

```python
# Correlation Matrix

correlation_matrix = (
    employee_dataframe[
        ["Age", "Salary"]
    ].corr()
)

correlation_matrix
```

```python
# Heatmap

sns.heatmap(
    correlation_matrix,
    annot=True
)

plt.show()
```
